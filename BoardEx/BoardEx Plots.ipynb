# Aggregate Ratio -- Options Excerised Spread Out Over Vesting Period

Options are exercised uniformly each period that they can be exercised. That is if an option is exercisable they exercise

*Ratio = Options Exercised / (Options Exercised + Exercisable Options + UnExercisable Options)* 

*Options Exercised = (1/length of vesting period) * (number of options package)*

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ratio_avg_agg = pd.read_csv('Clean/ratio_avg_agg_spread.csv', index_col=0)


plt.plot(range(2000,2021), ratio_avg_agg.T, marker='o')
plt.ylabel('Ratio')
plt.xlabel('Years')
plt.show()
```
